Pandas uses the plot() method to create diagrams.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')

df.plot()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')

df.plot(kind = 'scatter', x = 'SepalLengthCm', y = 'SepalWidthCm')

plt.show()
```
